# Fractals

```julia
# Julia set example from:
# http://www.laketide.com/julia-vs-r-vs-python/

# Set detail and resolution
detail = 1000
h = 6000
w = 6000

c = (φ-2) + (φ-1)im

function julia_fractal( z, c::Complex, l::Int64 )
    for i in 1:l
        abs(z) > 2 ? ( return i ) : ( z = z^2 + c )
    end
    return l
end

function generate_julia_set( h::Int64, w::Int64, c::Complex, l::Int64)
    bitmap = zeros(h,w)
    for x in 1:w, y in 1:h
        z = 3.*(x-w/2.)/w + (3.*(y-h/2.)/h)im # scale for proportion in Bitmap
        bitmap[y,x] = julia_fractal( z, c, l)/l
    end
    return bitmap
end
```

```julia
@time julia = generate_julia_set( h, w, c, detail)
```

```julia
# Mandelbrot set example

# Set detail and resolution
detail = 1000
h = 6000
w = 6000

c = (φ-2) + (φ-1)im

function mandelbrot_fractal( z, c::Complex, l::Int64 )
    x = 0. ; y = 0.
    x_temp = 0.
    for i in 1:l
        x^2. + y^2. > 4. ? ( return i ) : begin x_temp = x^2. - y^2. + real(z); y = 2.*x*y + imag(z); x = x_temp end
    end
    return l
end

function generate_mandelbrot_set( h::Int64, w::Int64, c::Complex, l::Int64)
    bitmap = zeros(h,w)
    for x in 1:w, y in 1:h
        z = ( -2.5 + 3.5*(x-1.)/(w-1.) ) + ( -1. + 2.*(y-1)/(h-1) )im # scale for proportion in Bitmap
        bitmap[y,x] = mandelbrot_fractal( z, c, l)/l
    end
    return bitmap
end
```

```julia
@time mandelbrot = generate_mandelbrot_set( h, w, c, detail)
```

```julia
Array{Gray24}( mandelbrot )
```

# Object Oriented Programming

```julia
type Data
  T::Array{Float64,1} # observation times
  δ::Array{Int64,1} # censoring indicator, 1 if exact observation
  Z₁::Array{Float64,1} # covariate in dimension 1
  Z₂::Array{Float64,1} # covariate in dimension 2
end
```

```julia
struct Model
    data::Data
    obs::Array{Float64,1} # observed times, sorted and without repetitions
    E::Array{Array{Int64,1},1} # array of indexes of exact observation corresponding to obs
    C::Array{Array{Int64,1},1} # array of indexes of censored observation corresponding to obs
    H₁ᵉ::Function # frequencies of exact observations in dimension
    H₂ᵉ::Function # frequencies of exact observations in dimension
    H₁ᶜ::Function # frequencies of censored observations in dimension 1
    H₂ᶜ::Function # frequencies of censored observations in dimension 1
    e_o::Array{Int64,1} # array of indexes of the exact observations in obs
    e_o_f::Function
end

# function to initialize model struct
function Model(data::Data)
    obs = sort( unique( data.T ) )
    E = Array{Int64,1}[]
    C = Array{Int64,1}[]
    for i in 1:length( obs )
        push!( E, intersect( find( data.T .== obs[i] ), find( data.δ .== 1 ) ) )
        push!( C, intersect( find( data.T .== obs[i] ), find( data.δ .== 0 ) ) )
    end
    h₁ᵉ(β₁::Float64) = [ mapreduce( j -> e^( β₁*data.Z₁[j] ), +, 0.,  E[i] ) for i in 1:length(obs)]
    h₂ᵉ(β₂::Float64) = [ mapreduce( j -> e^( β₂*data.Z₂[j] ), +, 0.,  E[i] ) for i in 1:length(obs)]
    h₁ᶜ(β₁::Float64) = [ mapreduce( j -> e^( β₁*data.Z₁[j] ), +, 0.,  C[i] ) for i in 1:length(obs)]
    h₂ᶜ(β₂::Float64) = [ mapreduce( j -> e^( β₂*data.Z₂[j] ), +, 0.,  C[i] ) for i in 1:length(obs)]
    H₁ᵉ(β₁::Float64) = [ cumsum( h₁ᵉ(β₁)[end:-1:1] )[end:-1:1]; 0]
    H₂ᵉ(β₂::Float64) = [ cumsum( h₂ᵉ(β₂)[end:-1:1] )[end:-1:1]; 0]
    H₁ᶜ(β₁::Float64) = [ cumsum( h₁ᶜ(β₁)[end:-1:1] )[end:-1:1]; 0]
    H₂ᶜ(β₂::Float64) = [ cumsum( h₂ᶜ(β₂)[end:-1:1] )[end:-1:1]; 0]
    e_o = find( E .!= [ [] ] )
    e_o_f(s) = intersect( find( obs .<= s ), e_o)
    return Model( data, obs, E, C, H₁ᵉ, H₂ᵉ, H₁ᶜ, H₂ᶜ, e_o, e_o_f )
end
```

# Parallel Computing

```julia
function compute_pi_vec( n::Int )
    x = rand(n) * 2 - 1 # uniformly distributed number on x-axis
    y = rand(n) * 2 - 1   # uniformly distributed number on y-axis
    r2 = x.^2 + y.^2  # radius squared, in radial coordinates
    return 4. * mean( 1.*(r2 .<= 1.0) )  
end
```

```julia
function compute_pi( n::Int )
    m = 0
    for i = 1:n
        x = rand() * 2 - 1 
        y = rand() * 2 - 1  
        r2 = x^2 + y^2 
        if r2 < 1.0
            m += 1
        end
    end

    return 4. * m/ n
end
```

```julia
@time compute_pi_vec(1)
```

```julia
@time compute_pi_vec(10^6)
```

```julia
x = collect(0:5000:1000000)
v = [ compute_pi( i ) for i in x];
```

```julia
scatter( x, v, legend = false)
xlabel!("Number of iterations")
title!( "Estimate of Pi")
```

```julia
@everywhere function compute_pi( n::Int )
    m = 0
    for i = 1:n
        x = rand() * 2 - 1 
        y = rand() * 2 - 1  
        r2 = x^2 + y^2 
        if r2 < 1.0
            m += 1
        end
    end

    return 4. * m/ n
end
```

```julia
job = @spawnat 6 compute_pi(1000000000)
```

```julia
function parallel_compute_pi( n::Int; c::Int=8)
    m = @parallel (+) for i = 1:c
        compute_pi(Int( n / c ))
    end

    return m / c
end
```

```julia
# initialize the attractor
n = 1500
dt = 0.02
σ, ρ, β = 10., 28., 8/3
x, y, z = 1., 1., 1.

# initialize a 3D plot with 1 empty series
plt = path3d(1, xlim=(-25,25), ylim=(-25,25), zlim=(0,50),
                xlab = "x", ylab = "y", zlab = "z",
                title = "Lorenz Attractor", marker = 1)
```

```julia
# build an animated gif, saving every 10th frame
@gif for i=1:n
    dx = σ*(y - x)     ; x += dt * dx
    dy = x*(ρ - z) - y ; y += dt * dy
    dz = x*y - β*z     ; z += dt * dz
    push!(plt, x, y, z)
end every 10
```

```julia
function doublependulum( du, state, p, t)
    G, L1, L2, M1, M2 = p
    du[1] = state[2]
    del_ = state[3] - state[1]
    den1 = (M1 + M2)*L1 - M2*L1*cos(del_)*cos(del_)
    du[2] = (M2*L1*state[2]*state[2]*sin(del_)*cos(del_) +
               M2*G*sin(state[3])*cos(del_) +
               M2*L2*state[4]*state[4]*sin(del_) -
               (M1 + M2)*G*sin(state[1]))/den1
    du[3] = state[4]
    den2 = (L2/L1)*den1
    du[4] = (-M2*L2*state[4]*state[4]*sin(del_)*cos(del_) +
               (M1 + M2)*G*sin(state[1])*cos(del_) -
               (M1 + M2)*L1*state[2]*state[2]*sin(del_) -
               (M1 + M2)*G*sin(state[3]))/den2
    return du 
end
```

```julia
u₀₁ = rand(4)
u₀₂ = u₀₁ + 0.1*randn(4)
tspan = (0.0, 10.0)
p =  [ G=10.0, L1 = 1.0, L2 = 1.0, M1 = 1.0, M2 = 1.0 ]
prob₁ = ODEProblem( doublependulum, u₀₁, tspan, p)
prob₂ = ODEProblem( doublependulum, u₀₂, tspan, p)
```

```julia
sol₁ = solve(prob₁,Tsit5(),reltol=1e-8,abstol=1e-8)
```
